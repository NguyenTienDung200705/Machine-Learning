### Bài 3:

### Bài 4:
Tạo hai mảng có kích thước (4, 4) chứa các số nguyên ngẫu nhiên từ 1 đến 10.

```python
arr4_1 = np.random.randint(1, 10, size = (4, 4))
arr4_2 = np.random.randint(1, 10, size = (4, 4))
print(arr4_1)
print("-----------")
print(arr4_2)
```

1. Tạo một mảng mask chứa giá trị True nếu phần tử lớn hơn 5 và False nếu
ngược lại.

```python
mask = arr4_1 > 5
mask1 = arr4_2 > 5
print(mask)
print("----------")
print(mask1)
```

2. Thực hiện phép nhân từng phần tử giữa hai mảng, nhưng chỉ nhân tại vị
trí có giá trị True trong mask.

```python
results = np.zeros_like(arr4_1)
results[mask] = arr4_1[mask] * arr4_2[mask]
print(results) 
```

3. Tính tổng từng hàng và từng cột của mảng kết quả.

```python
row_sum = results.sum(axis=1)
col_sum = results.sum(axis=0)

print("Hang : ", row_sum)
print("Cot : ", col_sum)
```

4. Chuẩn hóa từng hàng sao cho tổng mỗi hàng bằng 1.

```python
sum1 = np.where(row_sum == 0, 1 , row_sum)
normal = results / sum1[:, np.newaxis]

print(normal)
```

### Bài 5:
Tạo một mảng có 1000 phần tử lấy từ phân phối chuẩn (mean=0, std=1).

```python
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)
print(data)
```

1. Tính giá trị trung bình, phương sai, độ lệch chuẩn, tứ phân vị (Q1, Q2,
Q3) và IQR.

```python
mean = np.mean(data)
var = np.var(data)
std = np.std(data)

Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)
Q3 = np.percentile(data, 75)

IQR = Q3 - Q1

print("Mean : ", mean)
print("Phuong sai : ", var)
print("Do lech chuan : ", std)
print("Q1 : ", Q1, "Q2 : ", Q2, "Q3 : ", Q3)
print("Tu phan vi : ", IQR)
```

2. Phát hiện outliers sử dụng phương pháp IQR (Interquartile Range).

### Công thức
                       Lower bound = Q1 - 1,5 x IQR
                       Uper bound = Q3 + 1,5 x IQR

```python
lower_bound = Q1 - 1.5 * IQR
uper_bound = Q3 + 1.5 * IQR

outliers = data[(data < lower_bound) | (data > uper_bound)]

print("So luong outlier : ", len(outliers))
```

3. Vẽ histogram thể hiện phân phối dữ liệu.

```python
plt.hist(data, bins=30)
plt.title("Hisrogram")
plt.xlabel("Val")
plt.ylabel("Frequency")
plt.show()
```

4. Vẽ boxplot để kiểm tra sự tồn tại của outliers.

```python
plt.boxplot(data, vert=False)
plt.title("Boxplot of outliers")
plt.show()
```

5. Tạo một mảng mới đã loại bỏ outliers và so sánh phân phối trước & sau
khi lọc, sau đó:
- Vẽ histogram mới sau khi loại bỏ outliers.
- So sánh trung bình và độ lệch chuẩn của hai mảng.

```python
new_data = data[(data >= lower_bound) & (data <= uper_bound)]
print(new_data)
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(new_data, bins=30)
plt.title("Sau")

plt.subplot(1,2,2)
plt.hist(data, bins=30)
plt.title("Truoc")

plt.tight_layout()
plt.show()
```

```python
mean_new = np.mean(new_data)
std_new = np.std(new_data)

print("Mean truoc : ", mean, "Std truoc : ", std)
print("Mean sau : ", mean_new, "Std sau : ", std_new)
```
