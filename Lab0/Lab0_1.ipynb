### Bài 2: Viết chương trình nhập vào một số nguyên dương n.

1. Kiểm tra xem n có phải số nguyên tố không.

```python
n = int(input("Nhap"))
```

```python
if n < 2 :
    is_prime = False
else :
    is_prime = True
    for i in range(2, int(n**0.5)+1):
        if (n % i == 0):
            is_prime = False
            break
    if is_prime :
        print(n, "la so nguyen to")
    else :
        print(n, "khong phai so nguyen to")
```

2. In ra tất cả các ước số của n.

```python
for i in range(1, n+1):
    if (n % i == 0):
        print(i, end =" ")
```

3. Tìm tổng các chữ số của n.

```python
temp = n
tong = 0
while temp > 0:
    tong += temp % 10
    temp //= 10
print("tong cac chu so cua " ,n, "la", tong)
```
